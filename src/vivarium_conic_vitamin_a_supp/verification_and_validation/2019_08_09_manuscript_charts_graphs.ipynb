```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

# Simulation Visualizations
## Data

```python
years = np.array([2017, 2018, 2019, 2020, 2021, 2022, 2023])
nations = np.array(['Nigeria', 'Kenya', 'Burkina Faso'])
```

```python
averted_nigeria = np.array([29.717681836227694, 29.26669899234735, 262.33970957681987,
                            473.49711945446904, 668.1928909048063, 800.8905182380954,
                            1102.9324676333213])

unit_cost = 0.6
coverage_intervention_nigeria = np.array([0.32, 0.32, 0.388, 0.456, 0.524, 0.592, 0.66])
coverage_baseline_nigeria = np.array([0.32, 0.32, 0.32, 0.32, 0.32, 0.32, 0.32])
supplemented_nigeria = np.array([1283172.12, 1525432.71, 2138494.79, 2861839.16, 3668853.03,
                                 4586333.58, 5602350.94])
doses_intervention_nigeria = coverage_intervention_nigeria * supplemented_nigeria * 2
doses_baseline_nigeria = coverage_baseline_nigeria * supplemented_nigeria * 2
cost_intervention_nigeria = doses_intervention_nigeria * unit_cost
cost_baseline_nigeria = doses_baseline_nigeria * unit_cost
delta_cost_nigeria = cost_intervention_nigeria - cost_baseline_nigeria
icer_nigeria = delta_cost_nigeria / averted_nigeria

# total_intervention_cost_nigeria = 13642353.922271999
# total_baseline_cost_nigeria = 8319926.91072
total_icer_nigeria = delta_cost_nigeria.sum() / averted_nigeria.sum()
```

```python
averted_kenya = np.array([31.920396672166508, 31.835694821466532, 135.2700125775347,
                          159.45726876962775, 281.05684447373716, 427.04722052512034,
                          379.01471260580746])

unit_cost = 0.6
coverage_intervention_kenya = np.array([0.55, 0.55, 0.594, 0.638, 0.682, 0.726, 0.77])
coverage_baseline_kenya = np.array([0.55, 0.55, 0.55, 0.55, 0.55, 0.55, 0.55])
supplemented_kenya = np.array([2202727.9, 2611960.76, 3254149.51, 3973819.54, 4735836.2, 5574381.47,
                               6476317.15])
doses_intervention_kenya = coverage_intervention_kenya * supplemented_kenya * 2
doses_baseline_kenya = coverage_baseline_kenya * supplemented_kenya * 2
cost_intervention_kenya = doses_intervention_kenya * unit_cost
cost_baseline_kenya = doses_baseline_kenya * unit_cost
delta_cost_kenya = cost_intervention_kenya - cost_baseline_kenya
icer_kenya = delta_cost_kenya / averted_kenya

# total_intervention_cost_kenya = 23255935.055495996
# total_baseline_cost_kenya = 19027267.0698
total_icer_kenya = delta_cost_kenya.sum() / averted_kenya.sum()
```

```python
averted_burkina_faso = np.array([5.246980295264825, 5.157857716664438, 30.22658263429068,
                                 17.66428802627007, 58.597008256966454, 62.80426970003417,
                                 79.33522353457678])

unit_cost = 0.6
coverage_intervention_burkina_faso = np.array([0.88, 0.88, 0.892, 0.904, 0.916, 0.928, 0.94])
coverage_baseline_burkina_faso = np.array([0.88, 0.88, 0.88, 0.88, 0.88, 0.88, 0.88])
supplemented_burkina_faso = np.array([3530140.75, 4195756.5, 4882860.94, 5595671.835, 6285490.82, 
                                      7001417.465, 7724009.125])
doses_intervention_burkina_faso = coverage_intervention_burkina_faso * supplemented_burkina_faso * 2
doses_baseline_burkina_faso = coverage_baseline_burkina_faso * supplemented_burkina_faso * 2
cost_intervention_burkina_faso = doses_intervention_burkina_faso * unit_cost
cost_baseline_burkina_faso = doses_baseline_burkina_faso * unit_cost
delta_cost_burkina_faso = cost_intervention_burkina_faso - cost_baseline_burkina_faso
icer_burkina_faso = delta_cost_burkina_faso / averted_burkina_faso

# total_intervention_cost_burkina_faso = 42873818.944152
# total_baseline_cost_burkina_faso = 41411406.89135999
total_icer_burkina_faso = delta_cost_burkina_faso.sum() / averted_burkina_faso.sum()
```

## Graphs

```python
plt.figure(figsize=(6, 6))

plt.plot(years, supplemented_burkina_faso, label="Burkina Faso", c="g", linestyle = '-', marker = 'o')
plt.plot(years, supplemented_kenya, label="Kenya", c="r", linestyle = '-', marker = 'o')
plt.plot(years, supplemented_nigeria, label="Nigeria", c="b", linestyle = '-', marker = 'o')

plt.title('Supplemented Days')
plt.xlabel('Year')
plt.ylabel('Total Person-Days') # Ask Cody about person-years or days?
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6, 6))

plt.plot(years, averted_burkina_faso, label="Burkina Faso", c="g", linestyle = '-', marker = 'o')
plt.plot(years, averted_kenya, label="Kenya", c="r", linestyle = '-', marker = 'o')
plt.plot(years, averted_nigeria, label="Nigeria", c="b", linestyle = '-', marker = 'o')

plt.title('DALYs Averted')
plt.xlabel('Year')
plt.ylabel('Total Person-Years')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6, 6))

plt.plot(years, delta_cost_burkina_faso, label="Burkina Faso", c="g", linestyle = '-', marker = 'o')
plt.plot(years, delta_cost_kenya, label="Kenya", c="r", linestyle = '-', marker = 'o')
plt.plot(years, delta_cost_nigeria, label="Nigeria", c="b", linestyle = '-', marker = 'o')

plt.title('Additional Cost')
plt.xlabel('Year')
plt.ylabel('Total Person-Years')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6, 6))

plt.plot(years, icer_burkina_faso, label="Burkina Faso", c="g", linestyle = '-', marker = 'o')
plt.plot(years, icer_kenya, label="Kenya", c="r", linestyle = '-', marker = 'o')
plt.plot(years, icer_nigeria, label="Nigeria", c="b", linestyle = '-', marker = 'o')

plt.title('Incremental Costeffectiveness Ratio')
plt.xlabel('Year')
plt.ylabel('Total Person-Years')
plt.legend()

plt.show()
```
